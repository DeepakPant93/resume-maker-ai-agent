```python
import requests


def download_file(url, filename):
    # Send a HTTP GET request to the URL
    response = requests.get(url, stream=True)

    # Check if the request was successful
    if response.status_code == 200:
        # Open a local file with the specified filename in binary write mode
        with open(filename, 'wb') as file:
            # Write the content of the response to the file in chunks
            for chunk in response.iter_content(chunk_size=8192):
                file.write(chunk)
        print(f"File downloaded successfully as '{filename}'")
    else:
        print(f"Failed to download file. HTTP Status Code: {response.status_code}")

# URL of the media file to be downloaded
url = "https://aac.saavncdn.com/306/14fabd68de4410cee4b822c89f37abd8_160.mp4"

# Filename to save the downloaded file
filename = "downloaded_file.mp4"

# Call the function to download the file
download_file(url, filename)
```

```python
from urllib.parse import urljoin


def extract_html_and_urls(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Get the complete HTML content
        html_content = response.text
        print("Complete HTML content of the webpage:\n")
        print(html_content)
        ## write the html content in a file
        with open('html_content.txt', 'w') as file:
            file.write(html_content)

        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(html_content, 'html.parser')

        # Initialize lists to store URLs
        thumbnail_urls = []
        downloadable_urls = []

        # Extract internal URLs (e.g., thumbnail URLs and other downloadable resources)
        for img_tag in soup.find_all('img'):
            img_url = img_tag.get('src')
            if img_url:
                # Convert relative URLs to absolute URLs
                img_url = urljoin(url, img_url)
                thumbnail_urls.append(img_url)

        for a_tag in soup.find_all('a'):
            a_url = a_tag.get('href')
            if a_url and (a_url.endswith('.jpg') or a_url.endswith('.png') or a_url.endswith('.pdf') or a_url.endswith('.zip')):
                # Convert relative URLs to absolute URLs
                a_url = urljoin(url, a_url)
                downloadable_urls.append(a_url)

        # Print the extracted URLs
        print("\nThumbnail URLs:")
        for t_url in thumbnail_urls:
            print(t_url)

        print("\nDownloadable URLs:")
        for d_url in downloadable_urls:
            print(d_url)
    else:
        print(f"Failed to retrieve the webpage. HTTP Status Code: {response.status_code}")

# URL of the website to scrape
url = "https://www.jiosaavn.com/lyrics/ilahi-reprise-lyrics/QB4gUzt9blE"

# Call the function to extract HTML and URLs
extract_html_and_urls(url)
```

```python
from requests_html import HTMLSession


def fetch_rendered_html(url, render_timeout=20, retries=3):
    """
    Fetches and renders a webpage's HTML content after JavaScript execution.

    Args:
        url (str): The URL of the webpage.
        render_timeout (int): Timeout for JavaScript rendering.
        retries (int): Number of retries in case of failure.

    Returns:
        str: The rendered HTML content or an error message.
    """
    session = HTMLSession()

    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36"
    }
    session.headers.update(headers)

    for attempt in range(retries):
        try:
            # Fetch the webpage
            response = session.get(url, timeout=(10, 20))

            # Render the JavaScript
            response.html.render(timeout=render_timeout)

            # Return the rendered HTML content
            return response.html.html
        except Exception as e:
            print(f"Attempt {attempt + 1} failed: {e}")
        finally:
            session.close()

    return "Failed to fetch the rendered HTML after retries."

if __name__ == "__main__":
    url = "https://www.jiosaavn.com/song/kabira/MxEeByVxGnA"
    rendered_html = fetch_rendered_html(url)

    if rendered_html.startswith("<!DOCTYPE html"):
        print("Rendered HTML successfully fetched!")
    else:
        print("An error occurred.")

    # Save to file if needed
    with open("rendered_page.html", "w", encoding="utf-8") as file:
        file.write(rendered_html)
```

```python
import asyncio


async def fetch_page(url, render_timeout=20, retries=3):
    """
    Fetches and renders a webpage's HTML content asynchronously after JavaScript execution.

    Args:
        url (str): The URL of the webpage.
        render_timeout (int): Timeout for JavaScript rendering.
        retries (int): Number of retries for rendering.

    Returns:
        str: The rendered HTML content or an error message.
    """
    session = AsyncHTMLSession()

    for attempt in range(retries):
        try:
            # Fetch the webpage
            response = await session.get(url)

            # Render the JavaScript
            await response.html.arender(timeout=render_timeout)

            # Get the rendered HTML content
            rendered_html = response.html.html

            # Close the session
            await session.close()

            return rendered_html
        except Exception as e:
            print(f"Attempt {attempt + 1} failed: {e}")
        finally:
            await session.close()

    return "Failed to fetch the rendered HTML after retries."

if __name__ == "__main__":
    # Define the URL
    url = "https://www.jiosaavn.com/song/kabira/MxEeByVxGnA"

    # Run the asynchronous function
    try:
        rendered_html = asyncio.run(fetch_page(url))

        # Print or save the rendered HTML content
        if rendered_html.startswith("<!DOCTYPE html"):
            print("Rendered HTML successfully fetched!")
        else:
            print("An error occurred.")

        # Optionally save to a file for debugging
        with open("rendered_page.html", "w", encoding="utf-8") as file:
            file.write(rendered_html)
    except Exception as e:
        print(f"An error occurred during execution: {e}")
```

```python
from playwright.sync_api import sync_playwright


def scrape_dynamic_page(url, wait_time=5000):
    """
    Scrape a webpage including content loaded by JavaScript using Playwright
    
    Parameters:
    url (str): The URL to scrape
    wait_time (int): Maximum time to wait for dynamic content in milliseconds
    
    Returns:
    dict: Dictionary containing various elements from the page
    """
    with sync_playwright() as p:
        try:
            # Launch the browser
            browser = p.chromium.launch(headless=True)
            page = browser.new_page()

            # Go to the URL and wait for network to be idle
            page.goto(url, wait_until="networkidle")

            # Wait for the body to ensure page is loaded
            page.wait_for_selector("body", timeout=wait_time)

            # Collect elements
            elements = {
                "title": page.title(),
                "paragraphs": page.evaluate("""() => {
                    return Array.from(document.getElementsByTagName('p'))
                        .map(p => p.textContent.trim())
                        .filter(text => text.length > 0);
                }"""),
                "links": page.evaluate("""() => {
                    return Array.from(document.getElementsByTagName('a'))
                        .map(a => a.href);
                }"""),
                "buttons": page.evaluate("""() => {
                    return Array.from(document.getElementsByTagName('button'))
                        .map(b => b.textContent.trim())
                        .filter(text => text.length > 0);
                }"""),
                "divs": page.evaluate("""() => {
                    return Array.from(document.getElementsByTagName('div'))
                        .map(div => div.textContent.trim())
                        .filter(text => text.length > 0);
                }"""),
                # Get innerHTML of the entire body
                "full_html": page.evaluate("() => document.body.innerHTML")
            }

            browser.close()
            return elements

        except Exception as e:
            print(f"An error occurred: {e!s}")
            return None

def save_to_file(data, filename="output.json"):
    """Save the scraped data to a JSON file"""
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=2)

# Example usage
if __name__ == "__main__":
    # Replace with your target URL
    target_url = "https://example.com"

    # Scrape the page
    results = scrape_dynamic_page(target_url)

    if results:
        print("Page Title:", results["title"])
        print("\nParagraphs found:", len(results["paragraphs"]))
        print("\nLinks found:", len(results["links"]))
        print("\nButtons found:", len(results["buttons"]))

        # Save results to file
        save_to_file(results)
```

```python
from requests_html import HTMLSession


def scrape_page(url):
    """
    Scrape a webpage including JavaScript-rendered content
    using requests-html library
    """
    # Create an HTML Session object
    session = HTMLSession()

    try:
        # Get the page
        r = session.get(url)

        # Render JavaScript
        r.html.render(timeout=20)

        # Get elements
        elements = {
            'title': r.html.find('title', first=True).text if r.html.find('title') else '',
            'paragraphs': [p.text for p in r.html.find('p')],
            'links': [link for link in r.html.absolute_links],
            'divs': [div.text for div in r.html.find('div') if div.text],
            'html': r.html.html
        }

        return elements

    except Exception as e:
        print(f"An error occurred: {e!s}")
        return None
    finally:
        # Close the session
        session.close()

# Example usage
if __name__ == "__main__":
    url = "https://www.jiosaavn.com/song/kabira/MxEeByVxGnA"  # Replace with your URL
    data = scrape_page(url)

    if data:
        print("Title:", data['title'])
        print("\nNumber of paragraphs:", len(data['paragraphs']))
        print("\nNumber of links:", len(data['links']))

        # Print first few paragraphs as example
        print("\nFirst few paragraphs:")
        for p in data['paragraphs'][:3]:
            print(f"- {p[:100]}...")  # First 100 chars of each paragraph
```

```python
import asyncio
import time

import nest_asyncio
from pyppeteer import launch

# Enable nested event loops (fixes the runtime error)
nest_asyncio.apply()

async def scrape_page(url):
    """
    Scrape a webpage including JavaScript-rendered content
    """
    try:
        # Launch browser
        browser = await launch(headless=True, args=['--no-sandbox'])
        page = await browser.newPage()

        # Go to page and wait for content to load
        await page.goto(url, waitUntil='networkidle0')

        # Wait a bit for any dynamic content
        await page.waitFor(2000)

        # Get page content
        content = await page.content()

        # Parse with BeautifulSoup
        soup = BeautifulSoup(content, 'html.parser')

        # Get elements
        elements = {
            'title': soup.title.text if soup.title else '',
            'paragraphs': [p.text.strip() for p in soup.find_all('p') if p.text.strip()],
            'links': [urljoin(url, a.get('href')) for a in soup.find_all('a', href=True)],
            'divs': [div.text.strip() for div in soup.find_all('div') if div.text.strip()],
            'html': content
        }

        await browser.close()
        return elements

    except Exception as e:
        print(f"An error occurred: {e!s}")
        return None

def get_page_data(url):
    """
    Wrapper function to handle async call
    """
    return asyncio.get_event_loop().run_until_complete(scrape_page(url))

# Save data to file
def save_to_file(data, filename="output.json"):
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=2)

# Example usage
if __name__ == "__main__":
    url = "https://www.jiosaavn.com/song/kabira/MxEeByVxGnA"  # Your URL here
    data = get_page_data(url)

    if data:
        print("Title:", data['title'])
        print("\nNumber of paragraphs:", len(data['paragraphs']))
        print("\nNumber of links:", len(data['links']))

        # Save to file
        save_to_file(data)
```

```python

def setup_driver():
    # Setup Chrome WebDriver (or other driver)
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')
    driver = webdriver.Chrome(options=options)
    return driver

def get_downloadable_audio_link(url):
    if not url:
        return None

    # Extract the album ID and file ID from the URL
    parsed_url = urlparse(url)
    path_parts = parsed_url.path.split('/')
    album_id = path_parts[-2]
    file_id = path_parts[-1].split('.')[0]

    # Construct the downloadable audio link
    return f"https://aac.saavncdn.com/{album_id}/{file_id}.mp4"

def scrape_dynamic_page(url, wait_time=10):
    """
    Scrape a webpage including content loaded by JavaScript
    
    Parameters:
    url (str): The URL to scrape
    wait_time (int): Maximum time to wait for dynamic content to load
    
    Returns:
    dict: Dictionary containing various elements from the page
    """
    driver = setup_driver()

    try:
        # Load the page
        driver.get(url)

        # Wait for the button to be present
        button = WebDriverWait(driver, wait_time).until(
            EC.presence_of_element_located((
                By.CSS_SELECTOR,
                'a.c-btn.c-btn--primary[data-btn-icon="q"]'
            ))
        )

        # Click the button
        driver.execute_script("arguments[0].click();", button)

        # Wait a moment for any JavaScript updates
        time.sleep(2)

        # Get the updated HTML
        html_content = driver.page_source
        soup = BeautifulSoup(html_content, 'html.parser')

        # Extract elements
        elements = {
            'album_title': soup.title.text if soup.title else '',
            'description': soup.find('meta', {'name': 'description'})['content'] if soup.find('meta', {'name': 'description'}) else '',
            'album_title': soup.find('meta', {'property': 'og:title'})['content'] if soup.find('meta', {'property': 'og:title'}) else '',
            'album_description': soup.find('meta', {'property': 'og:description'})['content'] if soup.find('meta', {'property': 'og:description'}) else '',
            'album_url': soup.find('meta', {'property': 'music:album'})['content'] if soup.find('meta', {'property': 'music:album'}) else '',
            'album_image_url': soup.find('meta', {'property': 'twitter:image'})['content'] if soup.find('meta', {'property': 'twitter:image'}) else '',
            'song_info': {
                'name': soup.title.text if soup.title else '',
                'title': soup.find('meta', {'property': 'twitter:title'})['content'] if soup.find('meta', {'property': 'twitter:title'}) else '',
                'musician': [musician['content'] for musician in soup.find_all('meta', {'property': 'music:musician'})],
                'release_date': soup.find('meta', {'property': 'music:release_date'})['content'] if soup.find('meta', {'property': 'music:release_date'}) else '',
                'song_url': soup.find('meta', {'property': 'twitter:url'})['content'] if soup.find('meta', {'property': 'twitter:url'}) else '',
                'description': soup.find('meta', {'property': 'twitter:description'})['content'] if soup.find('meta', {'property': 'twitter:description'}) else '',
                'downloadable_url': get_downloadable_audio_link(soup.find('audio').find('source')['src'] if soup.find('audio').find('source') else ''),
                'song_lyrics_url': 'https://www.jiosaavn.com' + soup.find('a', title='Song Lyrics')['href'] if soup.find('a', title='Song Lyrics') else '',
            }
        }

        return elements

    except TimeoutException:
        print(f"Timeout waiting for page to load: {url}")
        return None
    except Exception as e:
        print(f"An error occurred: {e!s}")
        return None
    finally:
        driver.quit()

def scrape_multiple_pages(urls, wait_time=10):
    """
    Scrape multiple webpages and return a list of elements
    
    Parameters:
    urls (list of str): List of URLs to scrape
    wait_time (int): Maximum time to wait for dynamic content to load
    
    Returns:
    list of dict: List of dictionaries containing various elements from each page
    """
    results = []
    for url in urls:
        elements = scrape_dynamic_page(url, wait_time)
        if elements:
            results.append(elements)
    return results

# Example usage
urls = [
    "https://www.jiosaavn.com/song/kabira/MxEeByVxGnA",
    "https://www.jiosaavn.com/song/teri-mitti/SSUxeQxfX3o"
]

scraped_data = scrape_multiple_pages(urls)

# Save HTML to file (optional)
with open('elements.json', 'w', encoding='utf-8') as f:
    f.write(json.dumps(scraped_data, indent=2))
# for data in scraped_data:
#     print(data)
```
